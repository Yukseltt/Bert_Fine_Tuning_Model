```python
import re


input_file = "/content/drive/MyDrive/PDF_Library/metin_dosyasi.txt"
output_file = "/content/drive/MyDrive/PDF_Library/duzenlenmis_metin_dosyasi.txt"

def clean_text(text):
    # URL ve email temizleme - URL or email cleaning
    text = re.sub(r"http\S+|www\S+|@\S+", " ", text)

    # Fazla noktalama işaretlerini azaltma - Reducing excess punctuation
    text = re.sub(r"[;•·…<>|~=^_+*/\\]+", " ", text)

    # Fazla boşlukları temizleme - Reducing excess punctuation
    text = re.sub(r"\s+", " ", text)

    return text.strip()

with open(input_file, "r", encoding="utf-8", errors="ignore") as f:
    content = f.read()

cleaned = clean_text(content)

# Noktalara göre cümle bölme - Splitting sentences by periods
sentences = [s.strip() for s in re.split(r"[.!?]", cleaned)]

# Çok kısa olanları at (ör. <50 karakter) - Discard those that are too short (e.g. <50 characters)
sentences = [s for s in sentences if len(s) > 50]

# umcased eğitim için tüm karaketerleri küçük harfe dönüştür - umcased Convert all characters to lowercase for training
sentences = [s.lower() for s in sentences]

with open(output_file, "w", encoding="utf-8") as f:
    for s in sentences:
        f.write(s + ".\n")

print("Tamamlandı, toplam satır - Completed, total rows:", len(sentences))
print("Çıktı dosyası - Output file:", output_file)
```
