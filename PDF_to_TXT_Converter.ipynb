```python
import os
import fitz
import sys
from google.colab import drive


pdf_folder_path = '/content/drive/MyDrive/PDF_Library'


output_txt_file = '/content/drive/MyDrive/PDF_Library/metin_dosyasi.txt'


if not os.path.exists('/content/drive/MyDrive'):
    print("HATA: Google Drive bağlı değil - ERROR: Google Drive is not connected")
    print("Lütfen önce Google Drive bağlantı kodunu çalıştırın. - Pleasse run the Google Drive connection code first")
    sys.exit()


if not os.path.exists(pdf_folder_path):
    print(f"HATA: PDF klasörü bulunamadı - ERROR: PDF folder not found: '{pdf_folder_path}'")
    print("Lütfen 'pdf_folder_path' değişkenini doğru ayarladığınızdan emin olun - Please ensure that the 'pdf folder path' variable is set correctly.")
    sys.exit()


output_folder = os.path.dirname(output_txt_file)
if not os.path.exists(output_folder):
    print(f"Uyarı: Çıktı klasörü '{output_folder}' mevcut değil. Oluşturulmaya çalışılacak - Warning: Output folder '{output_folder}' does not exist. An attempt will be made to create it.")
    os.makedirs(output_folder, exist_ok=True)

print(f"'{pdf_folder_path}' klasörü taranıyor - folder is scanning...")

try:
    # Tüm PDF dosyalarını bul - Find all PDF files
    pdf_files = [f for f in os.listdir(pdf_folder_path) if f.lower().endswith('.pdf')]

    if not pdf_files:
        print(f"'{pdf_folder_path}' klasöründe hiç PDF dosyası bulunamadı - folder has not any PDF file")
    else:
        print(f"{len(pdf_files)} PDF dosyası bulundu - PDF files found")

        with open(output_txt_file, 'w', encoding='utf-8') as outfile:

            for i, filename in enumerate(pdf_files, 1):
                filepath = os.path.join(pdf_folder_path, filename)

                try:
                    print(f"[{i}/{len(pdf_files)}] İşleniyor - Processing: {filename}")

                    outfile.write(f"\n\n--- DOKÜMAN BAŞLANGICI - DOCUMENT BEGINNING: {filename} ---\n\n")

                    # PDF dosyasını aç - Open PDF file
                    doc = fitz.open(filepath)

                    # PDF'in tüm sayfalarını dolaş - Navigate all pages of a PDF
                    for page_num in range(len(doc)):
                        page = doc.load_page(page_num)
                        text = page.get_text() # Sayfadan metni çek - Pull text from page

                        # Metni temizle - Clear text
                        cleaned_text = " ".join(text.split())

                        outfile.write(cleaned_text + "\n")

                    doc.close()
                    outfile.write(f"\n--- DOKÜMAN SONU - END OF DOCUMENT: {filename} ---\n\n")

                except Exception as e:
                    print(f"HATA: {filename} dosyası işlenirken bir sorun oluştu: {e} - ERROR: There was a problem processing the file {filename}: {e}")
                    outfile.write(f"\n\n--- HATA: {filename} OKUNAMADI: {e} - ERROR: {filename} COULD NOT BE READ: {e} ---\n\n")

        print(f"\nİşlem tamamlandı - Process done")
        print(f"Tüm metinler başarıyla Google Drive'daki şu dosyaya kaydedildi - All texts have been successfully saved to the following file in Google Drive: '{output_txt_file}'")

        with open(output_txt_file, 'r', encoding='utf-8') as f:
            print("\nOluşturulan dosyanın içeriğinden bir önizleme - A preview of the contents of the generated file:")
            print(f.read(500) + "...")

except Exception as e:
    print(f"Beklenmedik bir hata oluştu - An unexpected error occurred: {e}")
```
